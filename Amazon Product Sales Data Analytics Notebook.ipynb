### Phase 4: Data cleaning (wrangling)

```python
# The fields 'Unnamed: 0', 'image', and 'link' are irrelevant to our analysis
# Let's remove these fields

df_dropped = df.drop(['Unnamed: 0', 'image', 'link'], axis = 1)

df_dropped
```

```python
# let's rename the columns to make them look proper.

df_dropped.columns = ['Name', 'Main Category', 'Sub-category', 'Ratings', 'No. Of Ratings', 'Discount Price', 'Actual Price']

df_dropped
```

```python
# Let's check for total number of missing values in each fields of the whole dataframe

df_dropped.isnull().sum()
```

```python
# Let's calculate the percentage of the total dataset that has missing values

print((df_dropped.isnull().sum().sum() / np.product(df_dropped.shape)) * 100)
```

```python
# Since the missing/NaN values amount to less than 25% (one-quarter) of the total dataset, we can safely remove these rows
# To drop all rows with NaN values or missing values

df_dropped_2 = df_dropped.dropna(axis=0)

df_dropped_2
```

```python
# Let's check for total number of missing values in each fields of the whole dataframe

df_dropped_2.isnull().sum()
```

```python
# Out of 551585 total rows, we have 340680 left after we removed missing/NaN values
# Let's see the datatypes of each of the fields in the dataframe

df_dropped_2.dtypes
```

```python
# We can see that even the Ratings, No. Of Ratings, Discount Price, and Actual Price fields are Object columns
# Let's change the datatypes to Decimal, Integer, Decimal, and Decimal respectively
```

```python
# to check for total number of missing values for ratings

df_dropped_2.Ratings.isnull().sum()
```

```python
# To check for unique values of ratings we used the unique() method.

df_dropped_2.Ratings.unique()
```

```python
# We cannot convert Ratings field from Object to float due to presence of non-numeric data
# We have to get rid of these non-numeric values, by using .str.contains method to filter them out then drop them
```

```python
# Lets match the rupee sign so that we can filter out irreleenat entries in our dataset.

df_dropped_2[df_dropped_2['Ratings'].str.contains('₹')]
```

```python
# droping the filter for rupee sign for which we used .str.contains method to filter out.

df_dropped_3 = df_dropped_2.drop(df_dropped_2[df_dropped_2['Ratings'].str.contains('₹')].index)

df_dropped_3 
```

```python
# Then we check again for unique values of ratings by using .unique method

df_dropped_3.Ratings.unique()
```

```python
# now let's filter out the Get & Free from this dataset using conditional filtering

df_dropped_3[(df_dropped_3.Ratings == 'Get') | (df_dropped_3.Ratings == 'FREE')]
```

```python
# After doing the conditional filtering for Get and free, now we can pass it into a new variable.

df_dropped_4 = df_dropped_3[(df_dropped_3.Ratings != 'Get') & (df_dropped_3.Ratings != 'FREE')]

df_dropped_4
```

```python
# Typecasting Ratings to Decimal so that we can perform the desired operations when needed.

df_dropped_4['Ratings'] = df_dropped_4['Ratings'].astype('float64')

df_dropped_4.dtypes
```

```python
# # To check for unique values for No of ratings 

df_dropped_4['No. Of Ratings'].unique()
```

```python
# We cannot convert No. Of Ratings field from Object to Integer when the records have comma sign
# We have to get rid of these comma signs, and replace them with empty string

df_dropped_4['No. Of Ratings'] = df_dropped_4['No. Of Ratings'].str.replace(',','')

df_dropped_4
```

```python
# remove all string values from a column using regex

df_dropped_4['No. Of Ratings'] = df_dropped_4['No. Of Ratings'].replace('[a-zA-Z]', np.nan, regex=True)

df_dropped_4
```

```python
## Typecasting No. Of Ratings to Integer (first to float because the column seems to have a mixture of integer and decimal)

df_dropped_4['No. Of Ratings'] = df_dropped_4['No. Of Ratings'].astype('float64')

df_dropped_4['No. Of Ratings'] = df_dropped_4['No. Of Ratings'].astype('int64')

df_dropped_4.dtypes
```

```python
# To check for unique values of discount price

df_dropped_4['Discount Price'].unique
```

```python
# We cannot convert Discount Price field from Object to Integer/Float when the records have comma sign and the rupee sign
# We have to get rid of these signs, and replace them with empty string

df_dropped_4['Discount Price'] = df_dropped_4['Discount Price'].str.replace(',','')

df_dropped_4['Discount Price'] = df_dropped_4['Discount Price'].str.replace('₹', '')
```

```python
### Typecasting Discount Price to Decimal so that we can perform the desired operations when needed.

df_dropped_4['Discount Price'] = df_dropped_4['Discount Price'].astype('float64')

df_dropped_4.dtypes
```

```python
#  To check for unique values of ratings 

df_dropped_4['Actual Price'].unique()
```

```python
# We cannot convert Actual Price field from Object to Integer/Float when the records have comma sign and the rupee sign
# We have to get rid of these signs, and replace them with empty string

df_dropped_4['Actual Price'] = df_dropped_4['Actual Price'].str.replace(',','')

df_dropped_4['Actual Price'] = df_dropped_4['Actual Price'].str.replace('₹','')
```

```python
### Typecasting  Actual Price to Integer/float so that we can perform the desired operations when needed.

df_dropped_4['Actual Price'] = df_dropped_4['Actual Price'].astype('float64')

df_dropped_4.dtypes
```

```python
### Lets look at our cleaned dataset so far...

df_dropped_5 = df_dropped_4

df_dropped_5
```

```python
# checking for duplicated values in our dataset

df_dropped_5.duplicated().any()
```

```python
# droping the duplicates in our dataset and passing it into a new variable.

df_dropped_5 = df_dropped_5.drop_duplicates()
```

```python
# After droping the duplicates, we can confirm if their is still duplicates.
# As we can see below it's False, meaning their is no more duplicates in our dataset.

df_dropped_5.duplicated().any()
```
